```python
# convert all dates under Value.dates.booked to datetime
data_transport['Value.dates.booked'] = pd.to_datetime(data_transport['Value.dates.booked'])
#print lowest and highest date
print(data_transport['Value.dates.booked'].min())
print(data_transport['Value.dates.booked'].max())
```

```python
def fetch_data(data, accountId, start_date, end_date):
    data = data[(data['Value.accountId'] == accountId) & (data['Value.dates.booked'] >= start_date) & (data['Value.dates.booked'] <= end_date)]
    return data
```
