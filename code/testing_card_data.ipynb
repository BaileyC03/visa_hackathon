```python

public_transport = ['PASSENGER RAILWAYS', 'BUS LINES', 'LOCAL COMMUTER TRANSPORT']
private_transport = ['PARKING LOTS,METERS,GARAGES', 'TAXICABS/LIMOUSINES', 'SERVICE STATIONS', 'COMMERCIAL EQUIPMENT - DEFAULT']
```

```python
# convert all the values in the cpd_dt column to datetime
transportation_data['cpd_dt'] = pd.to_datetime(transportation_data['cpd_dt'])
```

```python
# drop columns mrch_catg_cd	cp_flag	domestic_flag	intraregion_flag	interregion_flag	city_name	country_code	

transportation_data = transportation_data.drop(columns=['mrch_catg_cd', 'cp_flag', 'domestic_flag', 'intraregion_flag', 'interregion_flag', 'city_name', 'country_code'])
transportation_data.head()
```

```python
# define a function that will take start and end date , cluster_name_adjusted	and data 
# and return the total amount spent in the cluster between the start and end date on public transport and private transport

def get_total_spent(start_date, end_date, cluster_name_adjusted, data):
    # filter the data to include only the rows where the 'cluster_name_adjusted' column is equal to the cluster_name_adjusted parameter
    cluster_data = data[data['cluster_name_adjusted'] == cluster_name_adjusted]
    
    # filter the data to include only the rows where the 'cpd_dt' column is between the start_date and end_date parameters
    date_filtered_data = cluster_data[(cluster_data['cpd_dt'] >= start_date) & (cluster_data['cpd_dt'] <= end_date)]
    
    # filter the data to include only the rows where the 'mrch_catg_rlup_nm' column is in the public_transport list
    public_transport_data = date_filtered_data[date_filtered_data['mrch_catg_rlup_nm'].isin(public_transport)]
    
    # filter the data to include only the rows where the 'mrch_catg_rlup_nm' column is in the private_transport list
    private_transport_data = date_filtered_data[date_filtered_data['mrch_catg_rlup_nm'].isin(private_transport)]
    
    # calculate the total amount spent on public transport
    total_public_transport_spent = public_transport_data['spend'].sum()
    
    # calculate the total amount spent on private transport
    total_private_transport_spent = private_transport_data['spend'].sum()
    
    return total_public_transport_spent, total_private_transport_spent
```

```python
# print higest and lowest dates of the data
print(transportation_data['cpd_dt'].max())
print(transportation_data['cpd_dt'].min())
```

```python
start_date = '2023-05-01'
end_date = '2023-07-31'
cluster_name_adjusted = 'Family'

# call the get_total_spent function with the start_date, end_date, cluster_name_adjusted, and transportation_data as arguments
total_public_transport_spent, total_private_transport_spent = get_total_spent(start_date, end_date, cluster_name_adjusted, transportation_data)

# print the total amount spent on public transport and private transport
print('Total amount spent on public transport:', total_public_transport_spent)
print('Total amount spent on private transport:', total_private_transport_spent)
```

```python
# for each cluster, calculate the total amount spent on public transport and private transport between the start and end date and print

for cluster_name_adjusted in transportation_data['cluster_name_adjusted'].unique():
    total_public_transport_spent, total_private_transport_spent = get_total_spent(start_date, end_date, cluster_name_adjusted, transportation_data)
    print('Cluster:', cluster_name_adjusted)
    print('Total amount spent on public transport:', total_public_transport_spent)
    print('Total amount spent on private transport:', total_private_transport_spent)
    # ratio of public to private transport
    print('Ratio of public to private transport:', total_public_transport_spent/total_private_transport_spent)
    
    print()
```
